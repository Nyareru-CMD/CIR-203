```python
delivery_routes=["kisumu","nakuru","mombasa","nanyuki","siaya","busia","nyandarua","nairobi","kakamega","kitale"]
delivery_routes.append("kilifi")
print(delivery_routes)
delivery_routes.pop(2)
print(delivery_routes)
delivery_routes.sort()
print(delivery_routes)
delivery_routes.reverse()
print(delivery_routes)
```

```python
count_N = sum(route.startswith("n") for route in delivery_routes)
print("Routes starting with 'N':", count_N)
long_routes = [route for route in delivery_routes if len(route) > 5]
print("Long routes:", long_routes)
```
