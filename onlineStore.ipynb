```python
products={'bags':34,'shoes':7,'socks':54,'blankets':5,'shirts':6}
print(products)
products['jeans']=54
products['bags']=12
print(products)
del products['blankets']
print(products)
def low_stock_items(products):
    return [product for product, qty in products.items() if qty < 10]
print(low_stock_items(products))
for product, qty in products.items():
    print(f"{product}: {qty}")
```
