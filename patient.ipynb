```python
patient=('liam','67','121/80','75')
print(patient[0]+" is "+patient[1]+" years and has a heart rate of "+patient[3])
#its immutable(cant be changed) hence protect sensitive medical data
#its structured thus related data are grouped
#memory efficient
patient_list=list(patient)
print(patient_list)
patient_list[3]=70
print(patient_list)
patient=tuple(patient_list)
print(patient)
```
