```python
import numpy as np
transactions=np.array([[1240,6574,9674,7633,7643,9890],#branch 1
                      [3445,9708,5674,6740,6578,4352],#branch 2
                      [3450,9786,9765,8906,7890,6780],#branch 3
                      [1230,7890,4560,4570,4560,6750]])#branch 4
totalTransactionPerBranch=np.sum(transactions,axis=1)
print(totalTransactionPerBranch)
highest_transaction=np.argmax(totalTransactionPerBranch)
print(highest_transaction)
transactions_per_month=np.sum(transactions,axis=0)
print(transactions_per_month)
average_monthly_transaction=np.mean(transactions_per_month)
print(average_monthly_transaction)
```
